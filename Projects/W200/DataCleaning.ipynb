# Data Cleaning

This notebook contains the code that was used to clean the data that was downloaded from https://data.cityofchicago.org/api/views/ijzp-q8t2/rows.csv?accessType=DOWNLOAD. The file was downloaded as Crimes.csv in a local directory called Data. It was then split in multiple parts based on the quarters so that they could be uploaded to google dirve and then loaded into the notebook for processing.

```python
import pandas as pd
```

```python
crimes = pd.read_csv("../Data/Crimes.csv")
drop_cols =['Unnamed: 0','ID', 'Case Number', 'Updated On', 'Location', 'X Coordinate', 'Y Coordinate']
crimes.drop(drop_cols, axis=1, inplace=True)
```

```python
crimes['YrMth'] = pd.to_numeric(crimes['Date'].map(lambda x: x.strftime("%Y%m")))
```

```python
c2015Q1 = crimes[crimes['YrMth'] < 201504]
c2015Q2 = crimes[(crimes['YrMth'] > 201503) & (crimes['YrMth'] < 201507)]
c2015Q3 = crimes[(crimes['YrMth'] > 201506) & (crimes['YrMth'] < 201510)]
c2015Q4 = crimes[crimes['YrMth'] > 201509]
```

```python
c2016Q1 = crimes[crimes['YrMth'] < 201604]
c2016Q2 = crimes[(crimes['YrMth'] > 201603) & (crimes['YrMth'] < 201607)]
c2016Q3 = crimes[(crimes['YrMth'] > 201606) & (crimes['YrMth'] < 201610)]
c2016Q4 = crimes[crimes['YrMth'] > 201609]
```

```python
c2017Q1 = crimes[crimes['YrMth'] < 201704]
c2017Q2 = crimes[(crimes['YrMth'] > 201703) & (crimes['YrMth'] < 201707)]
c2017Q3 = crimes[(crimes['YrMth'] > 201706) & (crimes['YrMth'] < 201710)]
c2017Q4 = crimes[crimes['YrMth'] > 201709]
```

```python
c2015Q1.to_csv('../Data/Crimes2015Q1.csv')
c2015Q2.to_csv('../Data/Crimes2015Q2.csv')
c2015Q3.to_csv('../Data/Crimes2015Q3.csv')
c2015Q4.to_csv('../Data/Crimes2015Q4.csv')
c2016Q1.to_csv('../Data/Crimes2016Q1.csv')
c2016Q2.to_csv('../Data/Crimes2016Q2.csv')
c2016Q3.to_csv('../Data/Crimes2016Q3.csv')
c2016Q4.to_csv('../Data/Crimes2016Q4.csv')
c2017Q1.to_csv('../Data/Crimes2017Q1.csv')
c2017Q2.to_csv('../Data/Crimes2017Q2.csv')
c2017Q3.to_csv('../Data/Crimes2017Q3.csv')
c2017Q4.to_csv('../Data/Crimes2017Q4.csv')
```
